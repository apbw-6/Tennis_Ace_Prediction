## Returner position

Visualising returner position. As you can see below, some return positions seem strange. There are returns very close to the net, which is highly improbable. Even with the rare case of an underserve, it's hard to run up right to the net.

```python
fig, ax = tennis_utils.draw_tennis_court()

# Plot returner positions
sns.scatterplot(
    data=df, x="receiver_x", y="receiver_y", hue="serve_number", edgecolor="gray", s=8
)

# Aesthetics
ax.set_xlim(-22, 22)
ax.set_ylim(-10, 10)
ax.set_xticks([])
ax.set_yticks([])
# ax.set_xlabel("Court Length (m)")
# ax.set_ylabel("Court Width (m)")
ax.set_title("Returner Positions")

legend = ax.legend()
legend.set_title("Serve Number")
plt.legend(loc="lower right")
plt.show()
```

```python
sns.histplot(data=df, x="receiver_x", hue="serve_number", multiple="stack")
plt.title("Returner x-coordinate")
plt.show()

sns.histplot(data=df, x="receiver_y", hue="serve_number", multiple="stack")
plt.title("Returner y-coordinate")
plt.show()
```

## Ball bounce position

### Ball placement vs returner hand

Ball bouncing outside the service courts cannot be in. This dataset mentions that it is counting only serves that are called in. So these points have to be deleted.

```python
fig, ax = tennis_utils.draw_tennis_court()

# Plot ball bounce point (x,y)
sns.scatterplot(
    data=df,
    x="ball_bounce_x",
    y="ball_bounce_y",
    hue="receiver_hand",
    edgecolor="gray",
    s=8,
)

# Aesthetics
ax.set_xlim(-22, 22)
ax.set_ylim(-10, 10)
ax.set_xticks([])
ax.set_yticks([])
# ax.set_xlabel("Court Length (m)")
# ax.set_ylabel("Court Width (m)")
ax.set_title("Ball bounce positions during serve")

legend = ax.legend()
legend.set_title("Returner's hand")
plt.legend(loc="lower right")
plt.show()
```

### Ball placement for left handed receivers

I wanted to check ball placement for left handed receivers in the ad court. Since there are too many points for this kind of visualization to be useful, I took a sample to see if I can see clusters in the ball position. Indeed some clusters can be seen. Ball placement is near the T or near the sidelines, which is as expected. The clusters are also more dense towards the right side of the receiver, ie, the receiver is getting more backhands than forehands, which is also expected in tennis.

```python
df_left = df[df["receiver_hand"] == "left"]
```

```python
# There are too many points for visualisation to be useful, testing a sample dataset.
df_sampled = df_left.sample(n=1000, random_state=42)
```

```python
df_ad = df_sampled[df_sampled["serve_side"] == "ad"]
df_deuce = df_sampled[df_sampled["serve_side"] == "deuce"]
```

```python
fig, ax = tennis_utils.draw_tennis_court()

# Plot ball bounce point (x,y)
sns.scatterplot(
    data=df_ad,
    x="ball_bounce_x",
    y="ball_bounce_y",
    hue="receiver_hand",
    edgecolor="gray",
    s=8,
)

# Aesthetics
ax.set_xlim(-22, 22)
ax.set_ylim(-10, 10)
ax.set_xticks([])
ax.set_yticks([])
# ax.set_xlabel("Court Length (m)")
# ax.set_ylabel("Court Width (m)")
ax.set_title("Ball bounce positions on ad court")

legend = ax.legend()
legend.set_title("Returner's hand")
plt.legend(loc="lower right")
plt.show()
```
